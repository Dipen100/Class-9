## 2. Short Answer Questions:

### a. answer:-
- Computer hardware refers to the physical parts of the computer which we can see and touch. There are various types of computer hardware which is used for different operations like for input, output, storage. Example:- keyboard, mouse, monitor etc.

### b. answer:-
- The motherboard is also known as Printed Circuit Board (PCB), System Board or circuit board, serves as the central hub of a computer. It's like the backbone of the computer system where all components of computer system lies. It helps in connecting and faciliating communication between different parts of computer.

### c) answer:- 
- Microprocessor is the computing component which serves as the central processing unit (CPU) of a computer system. It gets data and instructions from other devices for processing. It does 4 operations:
    
    i. Retrieval/Fetch

    ii. Interpretation/Decode
    
    iii. Execution/Run
    
    iv. Record/Write Back

### d) answer:- 
- Computer memory is the place where all data and programs are stored in a computer. Mainly there are two types of memory:

**i. Primary memory**
- It is the main area of the computer system which holds data, instructions and programs temporarily in computer. It also has 2 types i.e RAM and ROM.

**ii. Secondary memory**
- It is the permanent memory of the computer system which holds data for the long period of time. The speed of secondary memory is slower than primary memory. It has many types like optical disks, magnetic disks, etc.

### e) answer:-
- Those devices which gives data and instructions to the computer are called input device. Data may be text, images, graphics and many more. Examples of input devices are keyboard, mouse, scanner, joystick etc.

### f) Answer:
- A computer mouse works by using a LED sensor which is placed at the bottm of the mouse which helps user to move in a particular direction. A left click is used to select files or folder, double left click opens it and right click is used to perform functions like rename, delete etc.

| **Mouse**                        | **Trackball**                           |
|-----------------------------|-------------------------------------|
| It uses LED sensor to sense direction. | It uses ball to change direction. |
| Entire mouse is moved.       | Only ball is rotated.              |
| It is lightweight and portable. | It is less portable.              |
| It is used on flat surfaces. | It can be used on any surface.     |

### g) answer:-
- A keyboard is an input device which is used to give text and numeric instructions to the computer system. It consists of numbers of key. They are:
    
    **i.  The alphanumeric key**
        - These keys contain keys for typing letters, numbers and special characters like punctuation marks.
    
    **ii. Modifier key**
        - These keys are not used alone. Each modifier key is used with another key.

    **iii. Numeric key**
        - These keys are located on the right side of the keyboard. It includes ten digits (0-9) and mathematical operations (+,-,*,/).
