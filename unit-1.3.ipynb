## 1. Very Short Answer Questions:
### a) answer:-
    - Computer Hardware refers to the physical parts of the computer which can be seen, felt and touch.

### b) answer:-
    - The motherboard is the platform to connect all the peripheral devices.

### c) answer:-
    - A CPU is the processing unit of computer system which process data to give meaningful output.

### d) answer:-
    - CPU stands for Central Processing Unit.

### e) answer:-
    - A keyboard is the input device which is used to give text and instruction to the computer system.

### f) answer:-
    - A mouse is an input device which is used to position the cursor on the screen.

### g) answer:- 
    - The primary function of mouse is to move the cursor on the screen, select objects and click on buttons.

### h) answer:-
    - RAM is the temporary memory of computer system which stores data and program until the computer has electrical power.

### i) answer:-
    - Harddrive is the primary storage device of the computer system which stores data and programs in it.

### j) answer:-
    - Any type of data can be stored on a hard drive like documents, images, videos, music and many more.

### k) answer:-
    - A graphic card is hardware component of computer system which is responsible for rendering and displaying images, videos, and animations on monitor.

### l) answer:- 
    - The purpose of graphic card is to accelerate the display of images and video on a computer screen.

### m) answer:-
    - PSU is the power supply unit of computer system which converts alternating current(AC) to direct current(DC).

### n) answer:-
    - PSU supplies direct current to the components of the computer system after converting from alternating current.

### o) answer:- 
    - A solid state drive (SSD) is the non-volatile storage media of computer which stores data until the power supply is on.

### p) answer:- 
    - A monitor is the softcopy output device which is used to display texts and graphics on the screen.

### q) answer:-
    - The main purpose of monitor in the computer system is to display texts and graphics on the screen.

### r) answer:-
    - A printer is the hardcopy output device that prints documents and images onto a paper.

### s) answer:- 
    - The purpose of printer in computer setup is to produce hardcopy output i.e permanent output which we can see and transfer easily.

### t) answer:- 
    - A scanner is an input device that inputs images, printed text, documents etc. to the computer system.

### u) answer:- 
    - In computing, scanner helps in converting physical objects into digital files that can be used on a computer.

### v) answer:-
    - USB flash drive is a portable storage device that connects to computers through a USB port.

### w) answer:-
    - An optical disk is a data storage medium that uses lasers to read and write information.

### x) answer:- 
    - A microphone is an input device that is used to input sound in a computer.

### y) answer:-
    - The primary function of microphone is to convert real sound wave into an electrical audio signal.

### z) answer:- 
    - A touchpad is small touch-sensitive pad used as an input device in laptop computers.

### aa) answer:- 
    - The function of touchpad in laptop is to move the pointer on the display screen.

### ab) answer:- 
    - A speaker is an output device that converts audio data to sound waves.
