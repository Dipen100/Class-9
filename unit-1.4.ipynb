## 1. Very short question answers.

##### a. Computer software is the collection of instructions that controls the operations of a computer system.

##### b. System software is the type of software that instructs a computer on how to use its components.

##### c. An example of system software is operating system like Microsoft Windows.

##### d. An operating system is a system software that serves as a master control program responsible for managing computer resources.

##### e. Utility software is a program or tool that performs specific tasks to enhance productivity, maintainance and operation of computer system.

##### f. Language processors are software tools that converts assembly-level language programs into machine code before executing them.

##### g. The different types of language processors are:
- *Assembler*

- *Compiler*

- *Interpreter*

##### h. The assembler are language processor that translates assembly language source code into machine code object programs and converting source program into object programs.

##### i. Any one feature of compiler is:
- *It converts high level language programs into machine code.*

##### j. Any one feature of interpreter is:
- *It converts one line of high level language into machine code immediately i.e it translates line by line.*

##### k. Any one difference between compiler and interpreter is: 
- *Compiler converts whole high level language into machine code directly but interpreter converts it line by line.*

##### l. Application software is a type of software that is designed to serve specific tasks for users.

##### m. An example of application software is package software like Microsoft Office.

##### n. A packaged software is a software that serves general tasks and is designed for use by broad range of users.

##### o. Any two features of packaged software are:
- *It is used by wide range of users.*

- *It is usually cheaper*

##### p. Any two features of tailored software are:
- *It is build to meet the specific requirements of a particular organization or user.*

- *It is more flexible and scalable.*

##### q. Word processing software is a computer program that is used to write, edit, format and print text documents.

##### r. An example of word processing software is Microsoft Word.

##### s. Spreadsheet software is a computer program that is used to organize, calculate and analyze data using rows and columns.

##### t. An example of spreadsheet software is Microsoft Excel.

##### u. The main purpose of spreadsheet software is to organize data in tables helping in budgeting, record keeping and data analysis.

##### v. Presentation software is a computer program used to create slides that display information using text, images, animations and videos.

##### w. An example of presentation software is Microsoft Powerpoint.

## 2. Short answer questions.


##### a. Software is the collection of programs that controls the operations of entire computer system. It differ from hardware in the follwing ways:

|Software|Hardware|
|---------|--------|
|*It is set of instructions.*|*It is physical part of computer.*|
|*It cannot be touched.*|*It can be touch and felt.*|
|*eg:- MS WORD, Windows etc.*|*eg:- Keyboard, Monitor, etc*|

##### b. System software and application software are the types of software. The main difference between them are:

| System Software | Application Software |
|-|-|
| *It is designed for managing system resources.* | *It is designed to serve specific tasks for users.* |
| *A computer cannot run without system software.* | *A computer can run easily without application software.* |
| *It doesn't depends on application software.* | *It depends on system software.* |
| *eg:- Windows.* | *eg:- MS Word.* |

##### c. The main purpose of operating system software is to control programs responsible for managing computer resources like input devices , processing units, memory units and output devices. It also acts as a bridge between application software and computer hardware. It's example is Microsoft Windows.

##### d. Utility software are designed to support the maintenance and operation of a computer system. The primary functions of utility software are:
- *It helps to transfer data, helps in data recovery, helps in virus scanning and removal.*

- *It also helps to sort programs.*

It's examples are antivirus programs, file management tools, text editor etc.

##### e. The difference between open-source software and propriety software with examples are:
| Open-source software | Propriety Software |
|-|-|
| *Source code is open and free.*| *Source code is closed and protected.* |
| *It can be modified and shared.* | *It cannot be modified and shared.* |
| *It is usually free of cost.* | *It is usually paid.* |
| *It is developed by communities.* | *It is developed by companies.* |
| *eg:- VLC, linux.* | *eg:- Windows, MS Word.* |

##### f. The two main categories of application software are:
- **1. Packaged Software**

    - *It is developed by software companies to serve general tasks and is designed for use by a broad range of users.*

    - *It's examples are spreadsheet applications, word processing software, etc.*

- **2. Tailored Software**

    - *It is developed to meet specific requirements and solve organization-specific problems.*

    - *It's examples are SLC result processing software, school management software, etc.*

##### g. Open-source software are free to use which are created by communities or developers. The source code of open-source software is freely available hence it can be used and modified by anyone. It is used for specific tasks or general purpose. The examples of open-source software are VLC media player and linux.
